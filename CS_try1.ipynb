### Question 1


## What is the distribution and total trip duration of males versus females for New York and Chicago, in minutes?

```r
# Use a "by command" to get the number of males versus females
by(ny$Trip.Duration/60,ny$Gender,sum)
by(chi$Trip.Duration/60,chi$Gender,sum)



```

```r
library("ggplot2")

```

```r
##Draw a histogram of different genders (NA, Female, Male) for New York and Chicago Data
ggplot (aes(x=Trip.Duration/60), data=subset(ny, !is.na (Gender)))+
    geom_histogram (binwidth=1, color = 'black', fill='#099DD9')+
    scale_x_continuous(breaks = seq(1,30,5), limits= c(0,30))+
    facet_wrap (~Gender)+
    ggtitle('Trip Duration in New York By Gender')+
    labs (x="Trip Duration in Minutes", y="Number of Users")

ggplot (aes(x=Trip.Duration/60), data=subset(chi, !is.na (Gender)))+
    geom_histogram (binwidth=1, color = 'black', fill='#099DD9')+
    scale_x_continuous(breaks = seq(1,30,5), limits= c(0,30))+
    facet_wrap (~Gender)+
    ggtitle('Trip Duration in Chicago By Gender')+
    labs (x="Trip Duration in Minutes", y="Number of Users")
```

The total numbers of minutes for females and males in NY is 177,517 and 476,741 minutes, respectively. The total numbers for Chicago were 22,227 and 56,355 for females and males, respectively. The histograms depect these differences and show that a common trip duration is approximately 6-7 minutes, for both genders.


### Question 2

# What is the median trip duration in hours for each city?

```r
#Return the median trip duration and ignore blanks.
median(chi$Trip.Duration/60, na.rm=TRUE)
median(ny$Trip.Duration/60, na.rm=TRUE)
median(wash$Trip.Duration/60, na.rm=TRUE)
```

```r
#Make boxplots of trip durations in each city.
library(ggplot2)
ggplot (aes(x=, y=Trip.Duration/60), data=chi)+
    geom_boxplot ()+
    coord_cartesian(ylim = c(0, 30))+
    ggtitle('Trip Duration in Chicago')+
    labs(x="Chicago", y="Duration in Minutes")
ggplot (aes(x=, y=Trip.Duration/60), data=ny)+
    geom_boxplot ()+
    coord_cartesian(ylim = c(0, 30))+
    ggtitle('Trip Duration in New York')+
    labs(x="New York", y="Duration in Minutes")
ggplot (aes(x=, y=Trip.Duration/60), data=wash)+
    geom_boxplot ()+
    coord_cartesian(ylim = c(0, 30))+
    ggtitle('Trip Duration in Washington, D.C.')+
    labs(x="Washington, D.C.", y="Duration in Minutes")
```

The median trip duration for Chicago, New York, and Washington, D.C. are 11.1, 10.1 and 11.7 minutes, respectively. These values and the distributions are also visualized in the box plots.

### Question 3

What are the counts of each user type for each city?

```r
#Get a Summary count of each user type in each data set.
summary(ny$User.Type)
summary(wash$User.Type)
summary(chi$User.Type)
```

```r
#Make bar graphs for user types in each city.
library("ggplot2")
ggplot(ny, aes(x= User.Type))+
    geom_bar(color='red', fill='#3399FF')+
    ggtitle('Number of Users By Type in New York')+
    labs(x="Type of User", y = "Number of Users")
ggplot(wash, aes(x= User.Type))+
    geom_bar(color='red', fill='#3344FF')+
    ggtitle('Number of Users By Type in Washington, D.C.')+
    labs(x="Type of User", y = "Number of Users")
ggplot(chi, aes(x= User.Type))+
    geom_bar(color='red', fill='#3377FF')+
    ggtitle('Number of Users By Type in Chicago')+
    labs(x="Type of User", y = "Number of Users")
```

For all cities, there were far more subscribers than customers for user types. In New York, there were 5,558 customers and 49,093 subscribers. In Washington, D.C. there were 23,450 customers and 65,600 subscribers. In Chicago, there were 1,746 customers and 6,883 subscribers.
